```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df
```

```python
#input output data
x = df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'Pedigree', 'Age']]
# x = df.drop('Outcome', axis = 1)
y = df['Outcome']
```

```python
sns.countplot(x = y)
```

```python
y.value_counts()
```

```python
# As we are using KNN algorithm which uses euclidean distance, we need to scale the data using scaler
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
```

```python
#cross validation
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, random_state=0, test_size=0.25)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, accuracy_score, classification_report
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
# recall for 1 is less
# lets find ideal value of k
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
error = []
for k in range (1, 41):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    pred = knn.predict(x_test)
    error.append(np.mean(pred!=y_test))
```

```python
plt.figure(figsize=(16, 9))
plt.xlabel('Value of k')
plt.ylabel('Error')
plt.grid()
plt.xticks(range(1, 41))
plt.plot(range(1, 41), error, marker = '.')
```

```python
knn = KNeighborsClassifier(n_neighbors=33)
```

```python
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
